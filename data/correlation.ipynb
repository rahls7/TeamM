```python
from scipy.stats.stats import spearmanr
import pandas as pd
```

```python
import matplotlib.pyplot as plt

# Plotting function to plot data points from Metric X and Metric Y
def plot(X, Y,xstring,ystring, sheetname, coefficient, value, proj):
    fig, ax1 = plt.subplots()
    ax1.plot(X, Y, 'bo')
    plt.xlabel(xstring)
    plt.ylabel(ystring)
    plt.title(r'Spearman Correlation'+' R='+str(coefficient)+' P='+str(value),fontweight="bold")
    plt.savefig(sheetname + proj+ '.png')
    plt.show()
```

```python
collection = pd.read_csv('./TestCoverage_Mutation_data/collection_merged.csv')
math = pd.read_csv('./TestCoverage_Mutation_data/math_merged.csv')
collection = collection.fillna(0)
math=math.fillna(0)
```

```python
#calculate correlation between statement and mutation
def corr_statement_mutation(df, proj):
  xstring='Statement_Percentage'
  ystring='Mutation_Score'
  X = df[xstring]
  Y = df[ystring]

  print(X)
  print(Y)
  coefficient, value = spearmanr(X, Y)
  coefficient= round(coefficient,2)
  value= round(value,4)
  print("correlation coefficient=", coefficient, "\n", "2 sided p value=",value)

  plot(X, Y,xstring,ystring, 'statement_mutation', coefficient, value, proj)
```

```python
corr_statement_mutation(collection, 'collection')
```

```python
corr_statement_mutation(math, 'math')
```
